```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
 ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList = [20,60,80,85,87,70,90,95,90,85]
```

```python
numAList = np.array(ageList)
numWList = np.array(weightList)
```

```python
numAList
```

```python
numWList
```

```python
plt.plot(numAList,numWList,"b")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age-weight graph")
#plt.show()
```

```python
numpArr1 = np.linspace(0,10,20)
```

```python
numpArr1
```

```python
numArr2 = numpArr1**3
```

```python
numArr2
```

```python
plt.plot(numpArr1,numArr2,"r--")
```

```python
plt.plot(numpArr1,numArr2,"y*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpArr1,numArr2,"b*-")
plt.subplot(1,2,2)
plt.plot(numArr2,numpArr1,"r--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numArr2,numpArr1,"g")
figureAxes.set_xlabel("x axis data")
figureAxes.set_ylabel("y axis data")
figureAxes.set_title("title")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.4,0.2,0.3,0.3])
eksen1.plot(numArr2,numpArr1,"g--")
eksen2.plot(numArr2,numpArr1,"y*-")
eksen1.set_title("first graph")
eksen2.set_title("second graph")
```

```python
(fFigure,fAxes) = plt.subplots(nrows=1,ncols=2)
type(fAxes)

for axes in fAxes:
    axes.plot(numpArr1,numArr2,"g")
    axes.set_xlabel("x axes")
plt.tight_layout()    
```

```python
newFigure = plt.figure(figsize=(9,7),dpi=300)
newAxes = newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxes.plot(numpArr1,numArr2, label="numpArr1 ** 3")
newAxes.plot(numpArr1,numpArr1**2,label="numpArr1 ** 2")
newAxes.legend(loc=2)
```

```python
newFigure.savefig("myFigure.png",dpi=500)
```
